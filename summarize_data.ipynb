## Summarization
Mean: the average.

Median: the middle value of the data. It divides the data into two halves, given that the data is arranged in some order, ascending or descending.

Mode: the most frequent element.

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import BinaryCrossentropy
```

```python
df = pd.read_csv('./zomato/zomato.csv')
df
```

```python
model = Sequential([
    Dense(units=1, activation='sigmoid'),
    Dense(units=5, activation='sigmoid'),
    Dense(units=1, activation='sigmoid')
])

model.compile(loss=BinaryCrossentropy())
X=df['rate (out of 5)']
Y=[]
for order in df['online_order']:
    if order == 'Yes': Y.append(1)
    else: Y.append(0)
Y = np.array(Y)
model.fit(X,Y,epochs=100)


```

```python
model.predict(np.arr[3.5])
```

```python
df = pd.DataFrame({"rating": [1, 2, 3, 4, 6, 7, 9, 9, 9, 10], "dummy": range(10)})
df
```

```python
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={'height_ratios' : (0.2,1)})

mean = df['rating'].mean()
median = df['rating'].median()
mode = df['rating'].mode().values[0]

sns.boxplot(data=df, x='rating', ax=ax_box)
ax_box.axvline(mean, color='r', linestyle='--')
ax_box.axvline(median, color='g', linestyle='-')
ax_box.axvline(mode, color='b', linestyle='-')

sns.histplot(data=df, x="rating", ax=ax_hist, kde=True)
ax_hist.axvline(mean, color='r', linestyle='--', label="Mean")
ax_hist.axvline(median, color='g', linestyle='-', label="Median")
ax_hist.axvline(mode, color='b', linestyle='-', label="Mode")

ax_hist.legend()

ax_box.set(xlabel='')
plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=100)

variance = np.var(data)

plt.figure(figsize=(10,6))

plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Example Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.axvline(np.mean(data), color='r', linestyle='dashed', lw=1, label='Mean')

plt.axvline(np.mean(data) - np.std(data), color='green', linestyle='dashed', linewidth=1, label='-1 Std Dev')
plt.axvline(np.mean(data) + np.std(data), color='green', linestyle='dashed', linewidth=1, label='+1 Std Dev')

plt.legend()
plt.grid()
plt.show()
```

```python
exam_scores = [85, 92, 78, 90, 88, 95, 85, 80, 82, 92]

variance = np.var(exam_scores)

plt.figure(figsize=(10, 6))

plt.hist(exam_scores, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Exam Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.axvline(np.mean(exam_scores), color='red', linestyle='dashed', linewidth=1, label='Mean')

plt.axvline(np.mean(exam_scores) - np.std(exam_scores), color='green', linestyle='dashed', linewidth=1, label='-1 Std Dev')
plt.axvline(np.mean(exam_scores) + np.std(exam_scores), color='green', linestyle='dashed', linewidth=1, label='+1 Std Dev')

plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.stats import pearsonr, spearmanr, kendalltau

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 6])

pearson_corr, _ = pearsonr(x,y)

spearman_corr, test = spearmanr(x,y)

kendall_corr, _ = kendalltau(x,y)

print(f"Pearson correlation: {pearson_corr:.2f}")
print(f"Spearman correlation: {spearman_corr:.2f}")
print(f"Kendall correlation: {kendall_corr:.2f}")
```

```python
periods = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
profits = [15000, 18000, 22000, 16000, 19000]

base_profit = profits[0]
profit_ratios = [profit / base_profit for profit in profits]

plt.figure(figsize=(10,6))

plt.bar(periods, profit_ratios, color='skyblue')
plt.title('Profit Ratio by Period')
plt.xlabel('Period')
plt.ylabel('Profit Ratio')
plt.ylim(0, max(profit_ratios) * 1.2)

plt.grid()
plt.show()
```
