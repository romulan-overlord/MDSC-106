```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data_dict = {
    't1' : [11,2,3,4],
    't2' : [11,22,33,44],
    't3' : [12,23,34,45],
    't4' : [21,32,43,54]
}

df = pd.DataFrame(data_dict)
```

```python
sns.heatmap(df, annot=True)
```

```python
sns.pairplot(df, kind='kde')
```

```python
sns.jointplot(df)
```
