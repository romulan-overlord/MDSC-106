```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = ['A', 'B', 'C', 'D']
y1 = [10,20,10,30]
y2 = [20,25,15,25]

plt.bar(x,y1,color='r')
plt.bar(x,y2,bottom=y1,color='b')
plt.legend(['Round 1', 'Round 2'])

plt.show()
```

```python
y3 = [10,15,20,5]
y4 = [20,20,15,25]

plt.bar(x,y1,color='r')
plt.bar(x,y2, bottom=y1,color='b')
plt.bar(x,y3, bottom=[i + j for i, j in zip(y1, y2)], color = 'g')
plt.bar(x,y4, bottom=[i + j + k for i, j, k in zip(y1, y2, y3)], color = 'y')

plt.show()
```
