```python
import numpy as np
from google.colab import autoviz

def scatter_plots(df, colname_pairs, figscale=1, alpha=.8):
  from matplotlib import pyplot as plt
  plt.figure(figsize=(len(colname_pairs) * 6 * figscale, 6 * figscale))
  for plot_i, (x_colname, y_colname) in enumerate(colname_pairs, start=1):
    ax = plt.subplot(1, len(colname_pairs), plot_i)
    df.plot(kind='scatter', x=x_colname, y=y_colname, s=(32 * figscale), alpha=alpha, ax=ax)
    ax.spines[['top', 'right',]].set_visible(False)
  plt.tight_layout()
  return autoviz.MplChart.from_current_mpl_state()

chart = scatter_plots(_df_8, *[[['Customer ID', 'Age'], ['Age', 'Purchase Amount (USD)'], ['Purchase Amount (USD)', 'Review Rating'], ['Review Rating', 'Previous Purchases']]], **{})
chart
```

```python
import numpy as np
from google.colab import autoviz

def heatmap(df, x_colname, y_colname, figscale=1, mpl_palette_name='viridis'):
  from matplotlib import pyplot as plt
  import seaborn as sns
  import pandas as pd
  plt.subplots(figsize=(8 * figscale, 8 * figscale))
  df_2dhist = pd.DataFrame({
      x_label: grp[y_colname].value_counts()
      for x_label, grp in df.groupby(x_colname)
  })
  sns.heatmap(df_2dhist, cmap=mpl_palette_name)
  plt.xlabel(x_colname)
  plt.ylabel(y_colname)
  return autoviz.MplChart.from_current_mpl_state()

chart = heatmap(_df_19, *['Color', 'Season'], **{})
chart
```

```python
# Filter the data for the specified 'Frequency of Purchases' values
selected_frequencies = ['Fortnightly', 'Weekly', 'Annually', 'Quarterly', 'Bi-Weekly', 'Monthly', 'Every 3 Months']
filtered_data = data[data['Frequency of Purchases'].isin(selected_frequencies)]

# Create a pivot table to prepare data for the stacked bar plot
pivot_data = pd.crosstab(index=filtered_data['Location'], columns=filtered_data['Frequency of Purchases'], normalize='index') * 100
plt.figure(figsize=(20, 10))
# Create a 100% stacked bar plot

sns.set_palette("husl")
pivot_data.plot(kind='bar', stacked=True)
plt.title("Frequency of Purchases by Location (100% Stacked)")
plt.xlabel("Location")
plt.ylabel("Percentage")
plt.xticks(rotation=90)
plt.legend(title='Frequency of Purchases', loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
color_likes = data.groupby(['Color', 'Gender'])['Review Rating'].apply(lambda x: (x >= 4).sum()).reset_index(name='Likes Count')

# Create a bar plot with Color on the x-axis
plt.figure(figsize=(12, 8))
sns.barplot(data=color_likes, x='Color', y='Likes Count', hue='Gender')
plt.title("Colors Liked by Gender")
plt.xlabel("Color")
plt.ylabel("Likes Count")
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

```python
# Group the data by 'Location', 'Shipping Type', and 'Item Purchased' and count the purchases
purchases_count = data.groupby(['Location', 'Shipping Type', 'Item Purchased']).size().reset_index(name='Purchase Count')

# Find the item with the highest purchase count for each Location and Shipping Type
max_purchases = purchases_count.groupby(['Location', 'Shipping Type'])['Purchase Count'].idxmax()

# Get the item with the highest purchase count for each Location and Shipping Type
most_purchased_items = purchases_count.loc[max_purchases]

# Create a bar plot
plt.figure(figsize=(40, 20))
sns.barplot(data=most_purchased_items, x='Location', y='Purchase Count', hue='Shipping Type')
plt.title("Most Purchased Item by Location and Shipping Type")
plt.xlabel("Location")
plt.ylabel("Purchase Count")
plt.xticks(rotation=45)
plt.legend(title='Shipping Type', loc='upper right')

# Enable hover tooltips
mplcursors.cursor(hover=True)

plt.show()
```

```python
# Group the data by 'Location' and 'Shipping Type' and count the purchases
shipping_type_counts = data.groupby(['Location', 'Shipping Type']).size().reset_index(name='Count')

# Find the location with the highest count for each Shipping Type
max_counts = shipping_type_counts.groupby('Shipping Type')['Count'].idxmax()

# Get the cities with the highest use for each Shipping Type
cities_max_use = shipping_type_counts.loc[max_counts]

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=cities_max_use, x='Shipping Type', y='Count', hue='Location')
plt.title("Cities with the Max Use of Each Shipping Type")
plt.xlabel("Shipping Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title='Location', loc='upper right')
plt.show()
```

```python
# Filter the data where Discount Applied is "Yes"
discounted_data = data[data['Discount Applied'] == "Yes"]

# Group the data by 'Category', 'Season', and 'Gender' and count the purchases
purchase_counts = discounted_data.groupby(['Category', 'Season', 'Gender']).size().reset_index(name='Purchase Count')

# Find the category with the highest purchase count for each combination of Season and Gender
max_purchases = purchase_counts.groupby(['Season', 'Gender'])['Purchase Count'].idxmax()

# Get the category with the highest purchase count for each Season and Gender
most_purchased_categories = purchase_counts.loc[max_purchases]

# Create a bar plot with Gender as hue
plt.figure(figsize=(12, 8))
sns.barplot(data=most_purchased_categories, x='Season', y='Purchase Count', hue='Gender', ci=None)
plt.title("Most Purchased Category with Discount Applied (Yes) by Season, by Gender")
plt.xlabel("Season")
plt.ylabel("Purchase Count")
plt.show()
```

```python
# Group the data by 'Item Purchased' and 'Gender' and count the promo code usage
promo_code_counts = data.groupby(['Item Purchased', 'Gender'])['Promo Code Used'].apply(lambda x: (x == 'Yes').sum()).reset_index(name='Promo Code Used Count')

# Create a bar plot with Gender as hue
plt.figure(figsize=(12, 8))
sns.barplot(data=promo_code_counts, x='Item Purchased', y='Promo Code Used Count', hue='Gender')
plt.title("Promo Code Usage by Item Purchased, with Gender as Hue")
plt.xlabel("Item Purchased")
plt.ylabel("Promo Code Used Count")
plt.xticks(rotation=90)
plt.legend(title='Gender')
plt.show()


```
